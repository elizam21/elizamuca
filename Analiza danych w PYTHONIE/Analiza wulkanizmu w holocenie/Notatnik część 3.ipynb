### Zadanie 4: Konstrukcja szeregów rodzielczych - bardziej skomplikowane tablice [10 pkt]

#### a) Forma tabelaryczna
Przygotuj szeregi rozdzielcze VEI w formie tablicy `vei_data` zawierającej następujące elementy:

> Indeks: VEI </br>
> Wartości VEI posortowane od najmniejszej do największej.

> Kolumny: 2-poziomowe nagłówki, dla każdego z nagłówków 1 poziomu powtarza się cały zestaw nagłówków 2 poziomu

>> Nagłówki 1 poziomu: All, Confirmed Eruption, Uncertain Eruption </br>
>> Całość danych oraz grupy wyznaczone na podstawie kolumny Eruption category</br>

>> Nagłówki 2 poziomu: f, cf, rf, crf </br>
>> Zliczenia poszczególnych kategorii (ang. frequency, absolute frequency), zliczenia skumulowane poszczególnych kategorii (ang. cumulative frequency, cumulative absolute frequency),  rf - częstość zliczeń poszczególnych kategorii (ang. relative frequency), crf - skumulowana częstość zliczeń poszczególnych kategorii (ang. cumulative relative frequency)

Wyświetl przygotowane tablicę w taki sposób, aby liczby w kolumnach f i cf oraz indeks tablicy VEI wyświetlały się bez miejsc po przecinku, a liczby z kolumn rf i crf z dokładnością do 4 miejsca po przecinku.

```python
# YOUR CODE HERE
def get_metrics(df: pd.DataFrame, category: str = None) -> pd.DataFrame:
    if category:
        df = df[df["Eruption Category"] == category]
    df = df.groupby("VEI").agg(f=("VEI", "count")) # frequency
    df["cf"] = df["f"].cumsum() # cumulative frequency
    df["rf"] = df["f"] / df["f"].sum() # relative frequency
    df["crf"] = df["rf"].cumsum() # cumulative relative frequency
    return df

all = get_metrics(eruptions_data)
confirmed = get_metrics(eruptions_data, category="Confirmed")
uncertain = get_metrics(eruptions_data, category="Uncertain")

vei_data = pd.concat([all, confirmed, uncertain], axis=1)
vei_data.index = vei_data.index.astype(int)
# vei_data = pd.pivot(all_data, columns=["Eruption Category"])
idx = pd.MultiIndex.from_product([['All eruptions', 'Confirmed Eruptions', 'Uncertain Eruptions'], ['f', 'cf', 'rf', 'crf']])
vei_data.columns = idx
vei_data.head()
```

#### b) Forma graficzna
Zwizualizuj dane z przygotowanej tablicy `vei_data` za pomocą wykresów słupkowych zliczeń (lewa oś wykresu) z nałożonymi na nie łamanymi częstości skumulowanej (prawa oś wykresu). Zestaw wykresy na pojedynczej figurze (siatka wykresów 1x3). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo ustaw:
- identyczne zakresy wartości na odpowiadających sobie osiach wszystkich trzech wykresów,
- różne od domyślnych i od siebie nawzajem kolory wykresu słupkowego oraz łamanej częstości,
- kolory lewej i prawej osi oraz ich opisów identyczne z kolorem elementu wykresu, którego dotyczą,
- format wyświetlania wartości VEI bez miejsc po przecinku.

```python
# YOUR CODE HERE
def plot_vei_distribution_v2(vei_data):
    fig, ax = plt.subplots(1, 3, figsize=(15, 5), facecolor="white")
    
    configs = [
        ('All eruptions', "All Eruptions"),
        ('Confirmed Eruptions', "Confirmed Eruptions"),
        ('Uncertain Eruptions', "Uncertain Eruptions")
    ]
    
    for i, (label, title) in enumerate(configs):
        max_frequency = vei_data[(label, 'f')].max() 
        ax[i].bar(vei_data.index, vei_data[(label, 'f')], color='darkgreen')
        ax[i].set_ylabel("f")
        ax[i].yaxis.label.set_color('darkgreen')
        ax[i].tick_params(axis='y', colors='darkgreen')
        ax[i].set_xlabel("VEI")
        ax[i].set_xticks(range(0, 8))
        ax[i].set_facecolor("white")
        
        ax_twin = ax[i].twinx()
        ax_twin.set_ylabel("crf")
        ax_twin.set_ylim([0, 1.1])
        ax_twin.yaxis.label.set_color('darkblue')
        ax_twin.tick_params(axis='y', colors='darkblue')
        ax_twin.plot(vei_data.index, vei_data[(label, 'crf')], c="darkblue")
        
        ax[i].spines['left'].set_color('darkgreen')
        ax_twin.spines['right'].set_color('darkblue')
        ax[i].set_title(title)
        ax[i].set_ylim(0,4500)
        ax[i].set_xlim([-0.5, 7.5])

    fig.suptitle("Volcano Eruption Distribution Charts", fontsize=16)
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.5)
    plt.show()

plot_vei_distribution_v2(vei_data)
```

### Zadanie 6: Wizualizacja rozkładu zarejestrowanych erupcji w czasie [3 pkt]

Korzystając z funkcji `seaborn.stripplot()` zwizualizuj jak rozkładały się w czasie wystąpienia zarejestrowanych erupcji. Do rozróżnienia erupcji w czasie wykorzystaj rok, w którym nastąpił ich początek.

Wyniki zestaw na pojedynczej figurze (siatka wykresów 2x1). Na górnym wykresie przedstaw dane rozróżnione tylko po kątem VEI, a na dolnym wykresie przedstaw dane rozróżnione zarówno po VEI jak i rodzaju dowodu ich wystąpienia (Evidence Type). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- dodaj legendę do dolnego wykresu w taki sposób, aby nie zasłaniała znajdujących nię na nim danych,

```python
# YOUR CODE HERE
fig, axs = plt.subplots(2, 1, figsize=(12, 10))
eruptions_data["VEI"] = eruptions_data["VEI"].astype(int)


sns.stripplot(data=eruptions_data, x='VEI', y='Start Year',alpha=0.6, size=4,jitter=True, ax=axs[0])
axs[0].set_title('Rozkład erupcji w czasie rozróżniony po VEI')
axs[0].set_facecolor("white")
axs[0].set_ylim(-12000, 2000)
axs[0].set_ylabel('Rok wybuchu')
axs[0].set_xlabel('VEI')

sns.stripplot(data=eruptions_data, x='VEI', y='Start Year', hue='Evidence Type',alpha=0.6,size=4, dodge=True, jitter=True, ax=axs[1])
axs[1].set_title('Rozkład erupcji w czasie rozróżniony po VEI i Evidence Type')
axs[1].set_facecolor("white")
axs[1].set_ylim(-12000, 2000)
axs[1].set_ylabel('Rok wybuchu')
axs[1].set_xlabel('VEI')

axs[1].legend(loc='upper left', bbox_to_anchor=(1, 1), title='Evidence Type')

plt.suptitle("Wizualizacja rozkładu zarejestrowanych erupcji w czasie", fontsize=16, fontweight='bold')

plt.tight_layout()

plt.show()
```

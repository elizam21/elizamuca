##### Wczytanie danych



```python
df = pd.read_csv("z_d_titanic.txt", skiprows=17, na_values='?', header=None) 
columns = ['pclass', 'survived', 'name', 'sex', 'age', 'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest']
df.columns=columns
df.head(20)
```

##### Wydzielamy kolumny z danymi jakościowymi i obliczamy ilość etykiet

```python
print(f"Liczba wszystkich pasażerów: {len(df.name.unique())}")
```

#### Duża liczebność cech:  
*ticket* - duża liczebność, część z biletów ma dodatkowo litery poprzedzające ID liczbowe, taka struktura utrudnia analizowanie, warto rozważyć np. wyodrębnienie prefiksów   
*cabin* -  jest to cecha bardzo szczegółowa, można spróbować wyodrębnić tylko litery z początku, może to mieć związek z lokalizacją kabiny względem bezpieczeństwa   
*boat* - wartość dla osób, które przeżyły, obecność lub nieobecność tej wartości niesie informację, można jej użyć np. by policzyć współczynnik przetrwania   
*home.dest* - również liczna cecha, często podawane jest to w schemacie miasto/kraj np. Paris/France, można spróbować zredukować tą cechę poprzez wydobycie informacji tylko o kraju, dzięki temu będzie łatwiej to sklasyfikować, a nie stracimy wiele informacji   

#### Mała liczebność cech: 
*pclass* - na Titanicu były 3 klasy podróżnicze, może być zakodowana jako zmienna kategoryczna  
*survived* - zmienna binarna 0/1, można ją bezpośrednio wykorzystać   
*sex* - cecha małoliczna, wiemy, że jest kobieta/mężczyzna , można zakodować jako zmienna 0/1   
*embarked* - mała liczba etykiet, (C = Cherbourg, Q = Queenstown, S = Southampton)     

```python
print(f"Wartość unikalnych etykiet dla zmiennej cabin: {len(df.cabin.unique())}")
```

#### Redukcja liczby cech dla zmiennej opisującej kabiny do etykiety zawierającej tylko pierwszą literę

```python
df["cabinReduced"] = df.cabin.astype(object).str[0]
display(df[["cabin", "cabinReduced"]].head(20))
```

##### Zmniejszenie kardynalności zbioru

```python
print('Liczba etykiet zmiennej cabinReduced: {}'.format(len(df.cabinReduced.unique())))
print(f'Kardynalność zbioru zmniejszyła się o { (len(df.cabin.unique()) - len(df.cabinReduced.unique())) / len(df.cabin.unique()) * 100:.2f}%')
```

#### CABIN
Dzięki powyższej operacji kardynalność zbioru zmniejszyła się o ponad 95%, powodem tego działania była duża granulacja zmiennej *cabin*, model stworozny w oparciu o jej pierwotną formę działałby niepoprawnie.
W wyniku tej operacji stracimy informację o dokładnym położeniu kabiny, natomiast zachowana zostanie wiedza o tym, na jakiej wysokości statku znajdowała się kabina. 
Zmniejszenie liczby unikalnych etykiet pozwala na lepszą interpretację wyników i minmalizuje szansę przeuczenia modelu. 
W wyniku takiej operacji utracimy szczegóły dotyczące kabin, np. nie będziemy wiedzieć, ile kabin posiadała dana rodzina, a także czy bliskość kabiny do dziury w statku wpływała na wzrost umieralności. 
Redukcja zmiennej *cabin* ma wpływ na uproszczenie i analizowanie modelu, natomiast trzeba być świadomym utraty części bardziej szczegółowych danych.

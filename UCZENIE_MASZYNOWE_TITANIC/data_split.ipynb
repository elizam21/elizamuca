```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("titanic_modified.csv") 
df.head(20)
```

### Zapoznanie się z funkcja  train_test_split() z biblioteki Scikit-learn
**sklearn.model_selection.train_test_split(arrays, test_size=None, train_size=None, random_state=None, shuffle=True, stratify=None)**

Funkcja **train_test_split()** służy do podziału danych na zbiór treningowy i testowy, zgodnie z podanymi przez nas parametrami. Przyjmuje jako argumenty ramki danych (DataFrame), tablice NumPy, listy lub macierze Scipy, które mają zostać rozdzielone na te dwa zbiory. Możemy określić rozmiar zbioru testowego lub treningowego — nie są to wartości z góry narzucone, co daje możliwość indywidualnego dostosowania proporcji podziału do konkretnego problemu.  

Parametr **random_state** pozwala uzyskać powtarzalność losowego podziału, **shuffle** decyduje o tym, czy dane mają zostać przemieszane przed podziałem, a **stratify** umożliwia zachowanie proporcji wybranej zmiennej w obu zbiorach. Dzięki temu możemy uniknąć sytuacji, w której pewne klasy są obecne tylko w zbiorze treningowym, a brakuje ich w testowym.

####  Tworzenie zmiennej z listą trzech kolumn: cabin, cabinReduced, sex

```python
col_name = ["cabin", "cabinReduced", "sex"]
df[col_name].head(20)
```

#### Dzielenie znioru na treningowy i testowy

```python
X = df[col_name]
y = df["survived"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

print(f"Rozmiar X_train {X_train.shape}, X_test {X_test.shape}, y_train {y_train.shape}, y_test {y_test.shape}")
```

Zmienna X zawiera 3 kolumny: **kabina**, **zredukowana kabina**, **płeć**.    
Zmienna y to nasza kolumna **survived**, czyli informacja, czy nasza osoba przeżyła (zmienna zależna).   
Podział danych został ustawiony 70% treningowych i 30% testowych, X_train i y_train mają 916 próbek, a X_test i y_test mają 393 próbki. 



#### Badanie rozkładu etykiet

```python
for zmienna in col_name:
  unique_test = [x for x in X_test[zmienna].unique() if x not in X_train[zmienna].unique()]
  print(f"{zmienna} ma {len(unique_test)} etykiet, które występują się w zbiorze testowym, a nie pojawiają się w treningowym.")
  print(unique_test)
```

W przypadku zmiennej **sex** obie etykiety występują zarówno w jednym, jak i w drugim zbiorze.  
Jeżeli chodzi o zmienną **cabin** to pojawia się 37 etykiet, które występują w zbiorze testowym, a nie ma ich w zbiorze treningowym, to znaczy, że model uczony na zbiorze treningowym nie miał okazji zobaczyć tych wartości *cabin* wcześniej, jest to zła informacja.  
W kwestii zmiennej **cabin_Reduced** tylko jedna unikalna etykieta (nan) występuje w zbiorze testowym, a nie w treningowym, wynika to z tego, że nan nie jest równe nan.

#### Kodowanie zmiennych kategorycznych do zmiennych liczbowych

```python
# Lista słowników do kodowania etykiet
encoding_dicts = []

# Dla każdej kolumny kategorycznej tworzymy słownik z przypisanymi wartościami liczbowymi
for column in col_name:
    label_to_int = {}
    for idx, label in enumerate(df[column].unique()):
        label_to_int[label] = idx
    encoding_dicts.append(label_to_int)

# Tworzenie nowych zakodowanych kolumn w zbiorach X_train i X_test
for idx in range(len(col_name)):
    col = col_name[idx]
    mapping = encoding_dicts[idx]
    
    X_train[f"{col}_encoded"] = X_train[col].map(mapping)
    X_test[f"{col}_encoded"] = X_test[col].map(mapping)

# Wyświetlenie przetworzonych zbiorów
display(X_train)
display(X_test)
```

#### Sprawdzanie liczby brakujących wartości w zmodyfikowanych zbiorach:

```python
for zmienna in col_name:
  print(f"Liczba NaN w kolumnie {zmienna}_map w zbiorze X_train: {X_train[zmienna+'_encoded'].isnull().sum()}, w X_test: {X_test[zmienna+'_encoded'].isnull().sum()}")
```

#### Komentarz na temat powyższych wyników:
Brak wartości NaN w kodowanych kolumnach świadczy o tym, że mapowanie przebiegło pomyślnie.   
Nie wystąpiły żadne nowe, niespodziewane wartości w zbiorze testowym, które nie byłyby znane w treningowym.   
Kodowanie zostało przeprowadzone przy użyciu słowników utworzonych na podstawie pełnego zbioru danych, co zapewniło, że etykiety z X_test również znalazły swoje odpowiedniki liczbowo zakodowane. Dzięki temu uniknięto powstawania NaN po .map()





#### Zastąpienie brakujących wartości liczbą 0

```python
encoded_cols = [col + '_encoded' for col in col_name]

X_train[encoded_cols] = X_train[encoded_cols].fillna(0)
X_test[encoded_cols] = X_test[encoded_cols].fillna(0)
```

Ma to sens w przypadku, gdy **0** jest traktowane jako oddzielna kategoria, czyli kategoria "braków", nasze zmienne cabin, sex i cabin_Reduced są jakościowe, nie ilościowe, gdy stałyby się ilościowe, a **0** byłoby wartością kodującą jakieś dane, to nie mogę tego zrobić.

#### Porównanie wartości unikalnych w zbiorze testowym i treningowym

```python
print(len(X_train.groupby(col_name).size()))
print(len(X_train))
print(len(X_test.groupby(col_name).size()))
print(len(X_test))
```

```python
print(len(X_train.groupby(col_name[1:]).size()))
print(len(X_train))
print(len(X_test.groupby(col_name[1:]).size()))
print(len(X_test))
```

#### Przed redukcją **cabin**
W zbiorze treningowym mamy 178 unikalnych kombinacji w kolumnach: **cabin, cabin_Reduced i sex**, co oznacza, że niektóre kombinacje pojawiają się wielokrotnie, ponieważ w zbiorze mamy 916 wierszy, wynika z tego, że większość cech w zbiorze jest powtarzalna.  
W zbiorze testowym mamy 77 unikalnych kombinacji, z tego wynika, że zbiór testowy zawiera mniej różnorodne przypadki niż treningowy zbiór danych. 
Jeśli większość kombinacji w zbiorze testowym jest obecna w zbiorze treningowym, model będzie miał szansę lepiej generalizować na testowy zbiór, ponieważ napotka te same kombinacje cech.
#### Po redukcji **cabin**
W zbiorze treningowym liczba unikalnych kombinacji zmniejszyła się do 14, a w zbiorze testowym do 12. 


Przy mapowaniu liczba etykiet pozostaje taka sama, a po redukcji etykiet jest mniej. Redukcja ma wpływ na działanie modelu i wyniki predykcji, natomiast mapowanie sprawia, że dane stają się łatwiejsze dla modelu. 

Zmniejszenie liczby unikalnych kategorii w cabin sprawia, że model nie musi radzić sobie z dużą liczbą różnych kategorii. To zmniejsza złożoność danych i zapobiega przeuczeniu (overfitting), ponieważ model nie musi zapamiętywać zbyt szczegółowych informacji.
